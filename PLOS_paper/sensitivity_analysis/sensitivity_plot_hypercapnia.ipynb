## Sensitivities with variable k_aut with healthy data

```python
df = pd.read_csv('hypercapnia/healthy/narrow_range/euclidean/sensitivities.txt', sep='\t')
```

```python
for p in (['u', 'P_a', 'SaO2sup', 'Pa_CO2']):#,"k_aut"]):    
    df = df.query("Parameter!='{}'".format(p))
df.shape
```

```python
def sensitivity(df, output):
    param = (output+"_mu_star", output+"_sigma")
    min_max_scaler = preprocessing.MinMaxScaler()
    x = df[param[0]].values.astype(float).reshape(-1, 1)
    # x = np.log(x+1)
    x_scaled = min_max_scaler.fit_transform(x)
    df_norm = pd.DataFrame({"Parameter": df["Parameter"].values, param[0]: x_scaled.flatten()})
    df_mu = df_norm.sort_values(by=param[0], ascending=False)
    
    df_sigma = df.sort_values(by=param[1], ascending=False)
    with sns.plotting_context("talk"):
        fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,10))
        sns.barplot(x=param[0], y="Parameter", data=df_mu[df_mu[param[0]]>0.2], ax=ax[0])
        sns.barplot(x=param[1], y="Parameter", data=df_sigma.head(10), ax=ax[1])
        ax[0].set_title(output + ": $\mu_{*}$")
        ax[1].set_title(output + ": $\sigma$")
        plt.tight_layout()
        plt.savefig('Figures/hypercapnia/healthy/narrow_range/euclidean/{}.png'.format(param))
        plt.show()
        plt.close()
    sensitive_parameters = df_norm[df_norm[param[0]]>0.2]["Parameter"].values
    
    return ax, df_norm, sensitive_parameters
```

```python
sens_params_euclidean = []
for o in ["CCO", "Vmca", "HHb", "HbO2"]:
    ax, df_norm, sens_p = sensitivity(df, o)
    sens_params_euclidean.extend(list(sens_p))
```
